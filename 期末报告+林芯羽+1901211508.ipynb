# 3. 实验
## 3.1准备实验数据集

```python
data = df.copy()
#针对分类变量设置虚拟变量
data = pd.get_dummies(data = data, columns =  ['hotel', 'meal', 'market_segment', 'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'customer_type', 'arrival_date_month','arrival_date_year'])

#agent数据因为本身便是用代码方式呈现，因此只需要对country变量做label Encoding
le = LabelEncoder()
data['country'] = le.fit_transform(data['country'])

#定义输入变量/输出变量并切分训练集及测试集
x = data.drop(['is_canceled'], axis=1).to_numpy().astype(np.float32)
y = data['is_canceled'].to_numpy().astype(np.int64)
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42)

#将输入变量做标准化处理
scaler = StandardScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train.shape
```

## 3.2 模型构建
### 基准：逻辑回归模型；对比模型：决策树模型、随机森林模型；主要模型：XGBoost模型、多层感知器神经网络模型

```python
#多层感知器神经网络模型网络架构
class MyNet(nn.Module):
    def __init__(self):
        super(MyNet, self).__init__()

        self.dense0 = nn.Linear(78, 128)
        self.bn0 = nn.BatchNorm1d(128)
        self.dropout = nn.Dropout(0.5)
        self.dense1 = nn.Linear(128, 64)
        self.bn1 = nn.BatchNorm1d(64)
        self.output = nn.Linear(64, 2)

    def forward(self, X, **kwargs):
        X = F.relu(self.dense0(X))
        X = self.bn0(X)
        X = self.dropout(X)
        X = F.relu(self.dense1(X))
        X = self.bn1(X)
        X = F.softmax(self.output(X), dim=-1)
        return X
```

```python
#设定各模型
base_models = [
                ('LR_model', LogisticRegression()),
                ('DT_model', DecisionTreeClassifier()),
                ('RF_model', RandomForestClassifier()),
                ('XGB_model', XGBClassifier()),
                ('NN_model', NeuralNetClassifier(MyNet, optimizer=torch.optim.Adam, max_epochs = 300, iterator_train__shuffle=True))]

#设定各模型参数组合
params = {
    'LR_model':{
        'C': [0.001, 0.01, 0.1, 1, 10, 100]
    },
    'DT_model':{
        'ccp_alpha': np.arange(0, 0.04, 0.01),
        'max_depth': [10, 13]
    },
    'RF_model': {
        'max_depth': [10, 13],
        'n_estimators': [100, 500],
        'min_samples_split': [2, 5]
    },
    'XGB_model':{
        'learning_rate':  [1e-3, 1e-2, 0.1],
        'max_depth': [5, 10],
        'n_estimators': [100, 500]
    },
    'NN_model':{
        'batch_size': [128, 200],
        'lr': [0.1, 0.01, 0.001]
    }
}

best_models = []

#使用5-fold交叉验证找出各模型最佳参数组合
for name, model in base_models:
    print(f'start fitting {name}')
    grid_cv = GridSearchCV(model,
                           params[name],
                           cv = 5)
    grid_cv.fit(x_train_scaled, y_train)
    best_models.append(grid_cv)
```

## 3.3 模型外样本精度测试

```python
#外样本预测精度测试
for i, model in enumerate(best_models):
    prediction = model.predict(x_test_scaled)
    accuracy = accuracy_score(y_test, prediction)
    precision = precision_score(y_test, prediction)
    recall = recall_score(y_test, prediction)
    f1_score = 2 * (precision * recall) / (precision + recall)
    print('{}最佳参数组合：{}, 最佳模型测试集accuracy：{:.2%} , precision：{:.2%} , recall：{:.2%}, f1-score：{:.2%} '.format(base_models[i][0], model.best_params_,  accuracy, precision, recall, f1_score))
```
